```python
import pandas as pd  

# Define file path with correct filename
file_path = r"C:\Users\user\Desktop\amazon sales analysis\amazon.csv"  

# Load data into pandas dataframe
df = pd.read_csv(file_path)

# Display first few rows
df.head()
```

```python
print(df.dtypes)
```

```python

# Remove ₹ symbol and commas from price columns, then convert to float
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')


# Convert discount percentage to float (remove % sign)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# Convert rating to float
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Fix rating_count (remove commas and convert to integer)
df['rating_count2'] = df['rating_count2'].str.replace(',', '').astype(float)

# Display updated data types
print(df.dtypes)

# Show first few rows to verify
df.head()
```

```python
print(df.dtypes)

```

```python
# Sort products by rating count in descending order
top_10_best_sellers = df.sort_values(by='rating_count2', ascending=False).head(10)

# Display top 10 best-selling products
print(top_10_best_sellers[['product_name', 'rating_count2', 'rating']])
```

```python
avg_discount = df["discount_percentage"].mean()
print(f"Average Discount: {avg_discount:.2f}%")
```

```python
# Calculate the correlation between the discount percentage and product rating
correlation = df["discount_percentage"].corr(df["rating"])
print(f"Correlation between discount and rating: {correlation:.2f}")  # Display the correlation coefficient
```

```python
# Find the product with the highest discount percentage
highest_discount = df.loc[df["discount_percentage"].idxmax(), ["product_name", "discount_percentage"]]
print(highest_discount)  # Display the product with the highest discount
```

```python
# Perform sentiment analysis on the review content using TextBlob
from textblob import TextBlob

def get_sentiment(text):
    # Return the sentiment polarity of the review text
    return TextBlob(text).sentiment.polarity

# Apply sentiment analysis to each review content
df['review_sentiment'] = df['review_content'].apply(get_sentiment)
print(df[['review_title', 'review_sentiment']].head())  # Display the review title and sentiment score
```

```python
# Visualize the distribution of actual and discounted prices using histograms
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
# Plot the distribution of actual prices
plt.hist(df["actual_price"], bins=30, alpha=0.7, label="Actual Price")
# Plot the distribution of discounted prices
plt.hist(df["discounted_price"], bins=30, alpha=0.7, label="Discounted Price")
# Add labels and a legend
plt.legend(loc='upper right')
plt.title("Price Distribution")
plt.show()
```
